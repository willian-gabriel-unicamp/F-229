```
def achar_pico(x,y,x_inicial,x_final):
  
  comecar = (list(x.loc[x == x_inicial].index))[0]
  terminar = (list(x.loc[x==x_final].index))[0]
  y_encontrado = np.max(y.loc[comecar:terminar])
  y_encon_ind = (list(y.loc[y==y_encontrado].index))[0]
  x_encontrado = x[y_encon_ind]
  return x_encontrado

def achar_vale(x,y,x_inicial,x_final):
  comecar = (list(x.loc[x == x_inicial].index))[0]
  terminar = (list(x.loc[x==x_final].index))[0]
  y_encontrado = np.min(y.loc[comecar:terminar])
  y_encon_ind = (list(y.loc[y==y_encontrado].index))[0]
  x_encontrado = x[y_encon_ind]
  return x_encontrado
```

# MMQ e análise do coeficiente angular

```
def mmq(x,y,ux=0,uy=0):
    """
    Calculates a linear fit using the methods of minium squares

    Parameters
    ----------
    x : np.ndarray 
        Array of x values.
    y : np.ndarray 
        Array of y values.
    ux : np.ndarray 
        Array of x uncertenties. The default is 0.
    uy : np.ndarray 
        Value or array of y uncertenties.The default is 0.

    Returns
    -------
    a : Float
        Slope.
    ua : Float
        Uncertenty in slope.
    b : Float
        Intercept.
    ub : Float
        Uncertenty in intercept.

    """
    if (type(uy) == float or type(uy) == int): #Incertezas iguais em y ou sem incertezas
        
        n = len(x)
        delta = n * np.sum(x*2) - np.sum(x)*2
      
        a = ( n * np.sum(x * y) - np.sum(x) * np.sum(y) )/delta
      
        b = (np.sum(y) * np.sum(x**2) - np.sum(x * y) * np.sum(x))/delta  
      
        ua= uy * np.sqrt(n/delta)
      
        ub= uy * np.sqrt(np.sum(x**2)/delta)
    
    elif type(uy) == np.ndarray and type(ux) == int: #Incertezas diferentes em y
    
        
        w = np.reciprocal(uy**2)
          
        delta = ( np.sum(w) * np.sum(w * x*2) ) - (np.sum(w * x))*2
          
        a = ( np.sum(w) * np.sum(w * x * y) - np.sum(w*x) * np.sum(w*y) )/delta
          
        b = (np.sum(w*y) * np.sum(w*x**2) - np.sum(w * x * y) * np.sum(w * x))/delta  
          
        ua=  np.sqrt(np.sum(w)/delta)
          
        ub= np.sqrt(np.sum(w * x**2)/delta)
          
    else: #Incertezas diferentes em x e y
    
             
        n = len(x)
        delta = n * np.sum((x*2)) - (np.sum(x))*2
        a = ( n * np.sum(x * y) - np.sum(x) * np.sum(y) )/delta #Aplica MMQ para uy, calcula a
      
        uy = np.sqrt(uy*2 + a2 * ux*2) #Projeta ux em uy
        
        w = uy**-2 #Aplica MMQ para o novo uy
      
        delta = ( np.sum(w) * np.sum(w * x*2) ) - (np.sum(w * x))*2
      
        a = ( np.sum(w) * np.sum(w * x * y) - np.sum(w*x) * np.sum(w*y) )/delta
      
        b = (np.sum(w*y) * np.sum(w*x**2) - np.sum(w * x * y) * np.sum(w * x))/delta  
      
        ua =  np.sqrt(np.sum(w)/delta)
      
        ub = np.sqrt(np.sum(w * x**2)/delta)
                
    return a,ua,b,ub
```
